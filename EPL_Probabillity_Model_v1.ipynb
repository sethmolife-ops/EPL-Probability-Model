```python
import requests
from bs4 import BeautifulSoup

# 1. Set the target URL for the Premier League data
url = "https://fbref.com/en/comps/9/Premier-League-Stats"

# 2. Disguise our script as a standard web browser to prevent getting blocked
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}

# 3. Ping the server and pull the raw HTML
print("Pinging FBref server...")
response = requests.get(url, headers=headers)
print(f"Server Status Code: {response.status_code} (200 means success!)")

# 4. Feed the raw HTML into BeautifulSoup to make it readable
soup = BeautifulSoup(response.text, 'html.parser')

# 5. Search the HTML for the main statistics table
standings_table = soup.find('table', class_='stats_table')

# 6. Print a small snippet to prove we captured it
if standings_table:
    print("\nTarget Acquired. Here is the raw HTML snippet:")
    print(standings_table.prettify()[:300])
else:
    print("Table not found. They may have updated their HTML tags.")
```

```python
import requests
import pandas as pd

# 1. Target URL and Browser Disguise
url = "https://fbref.com/en/comps/9/Premier-League-Stats"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64 AppleWebKit/537.36)'}

# 2. Ping the server
print("Pinging FBref...")
response = requests.get(url, headers=headers)
print(f"Status: {response.status_code}")

# 3. Deploy Pandas to forcefully extract every table on the page
try:
    # This single line replaces all the messy HTML parsing
    all_tables = pd.read_html(response.text)
    
    print(f"Extraction successful! Pandas isolated {len(all_tables)} tables from the raw code.")
    
    # 4. The main standings are always the first table on the page (Index 0)
    standings_df = all_tables[0]
    
    # 5. Print the first 5 rows of our new mathematical DataFrame
    print("\nTarget Acquired. Here is your structured data:")
    display(standings_df.head())
    
except ValueError:
    print("Pandas failed. FBref has deployed a Cloudflare blocker. We will need to pivot.")
```

```python
import pandas as pd

# 1. Target the raw CSV file for the 2023/2024 Premier League season (Complete data for backtesting)
csv_url = "https://www.football-data.co.uk/mmz4281/2324/E0.csv"

print("Bypassing HTML walls. Downloading raw quantitative CSV...")

# 2. Command Pandas to read the CSV directly into a DataFrame
epl_data = pd.read_csv(csv_url)

# 3. Success Check
print(f"Data secured! We have {epl_data.shape[0]} matches and {epl_data.shape[1]} statistical columns.")

# 4. Display the first 5 rows
display(epl_data.head())
```

```python
# 1. Isolate the signal from the noise
# We only want: Date, Teams, Goals, the Result (FTR), and the Bet365 Odds
target_columns = ['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'B365H', 'B365D', 'B365A']

# Create a clean DataFrame with just these columns
clean_data = epl_data[target_columns].copy()

# 2. FEATURE ENGINEERING (The Quant Secret)
# We need to convert the text 'H' (Home Win) into a binary mathematical integer (1 = Win, 0 = Not Win)
clean_data['Home_Win_Target'] = (clean_data['FTR'] == 'H').astype(int)

# 3. Print the optimized matrix
print("Data cleaned and Target Variable engineered. Look at the new column on the far right:")
display(clean_data.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 1. Define our Features (X) and our Target (y)
# We are asking the AI: "Based on the bookmaker odds (X), can you predict if the home team wins (y)?"
features = ['B365H', 'B365D', 'B365A']
X = clean_data[features]
y = clean_data['Home_Win_Target']

# 2. The Train/Test Split (Hiding 20% of the data)
# random_state=42 ensures we both get the exact same random split for debugging
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

print(f"Training data: {len(X_train)} matches. Hidden testing data: {len(X_test)} matches.")

# 3. Boot up the Mathematical Brain (Logistic Regression)
model = LogisticRegression()

# 4. Train the model (The AI is learning the patterns right now)
model.fit(X_train, y_train)

# 5. The Exam: Ask the AI to predict the 20% of matches it has never seen
predictions = model.predict(X_test)

# 6. Score the Exam
score = accuracy_score(y_test, predictions)
print(f"Initial AI Accuracy: {score * 100:.2f}%")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# --- PHASE 1: EXTRACTION ---
print("1. Extracting raw CSV data...")
csv_url = "https://www.football-data.co.uk/mmz4281/2324/E0.csv"
epl_data = pd.read_csv(csv_url)

# --- PHASE 2: ENGINEERING ---
print("2. Cleaning noise and engineering target variable...")
target_columns = ['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'B365H', 'B365D', 'B365A']
clean_data = epl_data[target_columns].copy()
# Create the binary 1 or 0
clean_data['Home_Win_Target'] = (clean_data['FTR'] == 'H').astype(int)

# --- PHASE 3: THE AI BRAIN ---
print("3. Executing Train/Test split...")
features = ['B365H', 'B365D', 'B365A']
X = clean_data[features]
y = clean_data['Home_Win_Target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

print("4. Training Logistic Regression algorithm...")
model = LogisticRegression()
model.fit(X_train, y_train)

print("5. AI is taking the exam on hidden data...")
predictions = model.predict(X_test)
score = accuracy_score(y_test, predictions)

print("--------------------------------------------------")
print(f"INITIAL AI ACCURACY: {score * 100:.2f}%")
print("--------------------------------------------------")
```
